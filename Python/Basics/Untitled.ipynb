```python
muskData = cv2.imread("musk.jpg")
```

```python
print(muskData.shape)
```

```python
plt.imshow(muskData)
```

```python
newMuskData =cv2.cvtColor(muskData, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(newMuskData)
```

```python
plt.figure(figsize=[20, 5])
plt.subplot(131); plt.imshow(newMuskData[:,:,0]); plt.title("Blue")
plt.subplot(132); plt.imshow(newMuskData[:,:,1]); plt.title("Green")
plt.subplot(133); plt.imshow(newMuskData[:,:,2]); plt.title("Red")
```
