## Contrast Vs Brightness Enhancement Comparison



```python
##########################################################################################
# Primary Contact    : v_s191@txstate.edu; vittal.siddaiah@gmail.com;
##########################################################################################
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
##########################################################################################

##########################################################################################
def Render(resultImageData, figId, titleStr):
    plt.subplot(figId)
    plt.imshow(resultImageData[..., ::-1])
    plt.xticks([]), plt.yticks([]) #To remove the ticks
    plt.title(titleStr)
##########################################################################################

##########################################################################################
def ContrastEnhance(imageData, pctgContrast): 
    if(pctgContrast > 1):
        pctgContrast /= 100
    pctgContrast += 1
    cntrstImage = (imageData * pctgContrast)/255
    clippImage = np.clip(cntrstImage,0,1)
    Render(clippImage, 132, "Contrast Enhanced Image")
##########################################################################################

##########################################################################################
def BrightEnhance(imageData, brightOffset): 
    bright = np.float32(imageData) + brightOffset
    brightNClipped = np.clip(bright/255,0,1)
    resultImage = np.uint8(brightNClipped*255)
    Render(resultImage, 133, "Brighness Enhanced Image")
##########################################################################################

def Enhance(imageData, offset):
    plt.figure(figsize=[20,10])
    Render(imageData, 131, "Original Image")
    BrightEnhance(imageData, offset)
    ContrastEnhance(imageData, offset)
##########################################################################################

 
imageData = cv2.imread("../../../Images/lenna-RGB.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")    
Enhance(imageData, 30)    


imageData = cv2.imread("../../../Images/fingerprint.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")    
Enhance(imageData, 40)    

imageData = cv2.imread("../../../Images/woman.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")    
Enhance(imageData, 50)    
```
