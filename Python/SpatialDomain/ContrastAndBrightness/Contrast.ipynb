## Contrast Enhancement

```python
##########################################################################################
# Primary Contact    : v_s191@txstate.edu; vittal.siddaiah@gmail.com;
##########################################################################################
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
##########################################################################################

##########################################################################################
def Render(resultImageData, figId, titleStr):
    plt.subplot(figId)
    plt.imshow(resultImageData[..., ::-1])
    plt.xticks([]), plt.yticks([]) #To remove the ticks
    plt.title(titleStr)
##########################################################################################

##########################################################################################
def ContrastEnhance(imageData, pctgContrast): 
    plt.figure(figsize=[20,10]) 
    Render(imageData, 121, "Original Image")
    if(pctgContrast > 1):
        pctgContrast /= 100
    pctgContrast += 1
    cntrstImage = (imageData * pctgContrast)/255
    clippImage = np.clip(cntrstImage,0,1)
    Render(clippImage, 122, "Contrast Enhanced Image")
##########################################################################################

##########################################################################################
imageData = cv2.imread("../../../Images/lenna-RGB.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")    
ContrastEnhance(imageData, 30)    


imageData = cv2.imread("../../../Images/fingerprint.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")    
ContrastEnhance(imageData, 30)    

imageData = cv2.imread("../../../Images/chestXray2.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")    
ContrastEnhance(imageData, 20)    
    
```
