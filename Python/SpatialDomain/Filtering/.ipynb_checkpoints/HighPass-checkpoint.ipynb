## High Pass Filter / Image Sharpening

### Steps are:
1. Apply a low pass filter on an image.
2. Obtain the high-frequency information by subtracting the above low pass image from the original image.
3. Apply the above high-frequency information into the original image with a control parameter


```python
##########################################################################################
# Primary Contact    : v_s191@txstate.edu; vittal.siddaiah@gmail.com;
##########################################################################################
#Image Filtering using Convolution
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def Render(resultImageData, figId, titleStr):
    resultImageData = cv2.convertScaleAbs(resultImageData)
    plt.subplot(figId)
    plt.imshow(resultImageData[..., ::-1])
    plt.xticks([]), plt.yticks([]) #To remove the ticks
    plt.title(titleStr)
    

def HighPassFilter(imageData):
    plt.figure(figsize=[20,10])
    
    Render(imageData,121,"Original Image")

    sharpenKernel = np.array((
        [0, -1, 0],
        [-1, 5, -1],
        [0, -1, 0]), dtype="int")

    # Applying filter2D with the sharpening kernel
    sharpenOutput = cv2.filter2D(imageData, -1, sharpenKernel)
    Render(sharpenOutput,122,"Sharpened Result")

    
imageData = cv2.imread('../../../Images/barbara.tif')
HighPassFilter(imageData)

imageData = cv2.imread('../../../Images/girl-blurred.tif')    
HighPassFilter(imageData)    
```

```python
imageData = cv2.imread('../../Images/lenna-RGB.tif')    
HighPassFilter(imageData)    


imageData = cv2.imread('../../Images/pollen-lowcontrast.tif')    
HighPassFilter(imageData)    
```
