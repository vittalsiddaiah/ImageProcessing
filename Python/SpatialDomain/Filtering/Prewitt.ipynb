## Prewitt Filter

Prewitt filter is based on the first derivative, to make gradient calculations.


### Function Syntax
I could not find a Prewitt filter pre-defined in OpenCV so deriving the filter from the fundamentals.



```python
##########################################################################################
# Primary Contact    : v_s191@txstate.edu; vittal.siddaiah@gmail.com;
##########################################################################################
#Image Filtering using Convolution
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

################################################################################
def Render(resultImageData, figId, titleStr):
    resultImageData = cv2.convertScaleAbs(resultImageData)
    plt.subplot(figId)
    plt.imshow(resultImageData[..., ::-1])
    plt.xticks([]), plt.yticks([]) #To remove the ticks
    plt.title(titleStr)
    
################################################################################

################################################################################
#prewitt
def PrewittFilter(imageData):
    kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
    kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    print(kernelx)
    print(kernely)
    
    plt.figure(figsize=[20,10]) 
    Render(imageData, 131, "Original")

    imagePrewittX = cv2.filter2D(imageData, -1, kernelx)
    Render(imagePrewittX, 132, "Prewitt with X Gradient")
    
    imagePrewittY = cv2.filter2D(imageData, -1, kernely)
    Render(imagePrewittY, 133, "Prewitt with Y Gradient")
    return
################################################################################


imageData = cv2.imread('../../../Images/barbara.tif')
PrewittFilter(imageData)


```
