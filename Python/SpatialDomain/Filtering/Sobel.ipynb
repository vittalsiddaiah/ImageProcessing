## Sobel Filter

Sobel filter is based on the first derivative, to make gradient calculations even more robust and free of noise.


### Function Syntax
``` python
	dst =   cv2.Sobel(  src, ddepth, dx, dy[, dst[, ksize[, scale[, delta[, borderType]]]]] )
```

Parameters
* **`src`** Source image.
* **`dst`** Destination image of the same size and the same number of channels as src .
* **`ddepth`**	Desired depth of the destination image.
* **`dx`** order of the derivative x.
* **`dy`** order of the derivative y.
* **`ksize`**	Sobel kernel size [1, 3, 5, 7]
* **`scale`**	Optional scale factor for the computed Laplacian values. By default, no scaling is applied. 
* **`delta`**	Optional delta value that is added to the results prior to storing them in dst .
* **`borderType`**	Pixel extrapolation method, see BorderTypes. BORDER_WRAP is not supported.

[cv.Laplacian](https://docs.opencv.org/3.4/d4/d86/group__imgproc__filter.html#gad78703e4c8fe703d479c1860d76429e6)

````python
##########################################################################################
# Primary Contact    : v_s191@txstate.edu; vittal.siddaiah@gmail.com;
##########################################################################################
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def Render(resultImageData, figId, titleStr):
    resultImageData = cv2.convertScaleAbs(resultImageData)
    plt.subplot(figId)
    plt.imshow(resultImageData[..., ::-1])
    plt.xticks([]), plt.yticks([]) #To remove the ticks
    plt.title(titleStr)
    
def SobelFilter(imageData):
    for kernelSize in [3, 5]:
        plt.figure(figsize=[20,10]) 

        Render(imageData, 131, "Original Image")

        resultImageData = cv2.Sobel(imageData, cv2.CV_64F, 1, 0, kernelSize)
        titleStr = "Sobel with X Gradients Kernel = %d "% kernelSize
        Render(resultImageData, 132, titleStr)

        resultImageData = cv2.Sobel(imageData, cv2.CV_64F, 0, 1, kernelSize)
        titleStr = "Sobel with Y Gradients Kernel = %d "% kernelSize
        Render(resultImageData, 133, titleStr)

######################################################################        

######################################################################
imageData = cv2.imread("../../../Images/building-cropped-834by1114.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")
SobelFilter(imageData)
######################################################################
imageData = cv2.imread("../../../Images/barbara.tif")  #Color Image is Scanned
if imageData is None: print("Unable to open the image! please verify !!!")
SobelFilter(imageData)
######################################################################
````
